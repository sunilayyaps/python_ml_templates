```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data = pd.read_csv('D:/data/50_Startups.csv')
data.head()
```

```python
pd.tools.plotting.scatter_matrix(data)
plt.show()
```

```python
X = data.iloc[:, :-1].values
Y = data.iloc[:, -1].values
print(X.shape,Y.shape)
```

```python
#labelencoder turns each label into its own number in the same column
labelencoder = LabelEncoder()
X[:, 3] = labelencoder.fit_transform(X[:,3])
print('Before OneHotEncoder: {0}'.format(X.shape))
#onehotencoder takes the categories converted by label encoder and breaks them out into new colums
onehotencoder = OneHotEncoder(categorical_features=[3])
X = onehotencoder.fit_transform(X).toarray()
print('After OneHotEncoder: {0}'.format(X.shape))
#to avoid the dummy variable trap, exclude one of the dummy variables
X = X[:, 1:]
print('After removing a dummy variable: {0}'.format(X.shape))
```

```python
#split the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#now fit the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)
#make prediciton on test set
y_pred = model.predict(X_test)
```

```python
plt.scatter(y_pred, Y_test)
plt.show()
```
