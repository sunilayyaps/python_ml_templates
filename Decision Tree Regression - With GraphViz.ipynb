```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings(action='ignore', module='scipy', message='^internal gelsd')
```

```python
dataset = pd.read_csv('data/Position_Salaries.csv')
dataset.head()
```

```python
features = ['Level']
target = ['Salary']
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(random_state=0)
dt_reg.fit(dataset[features], dataset[target])
```

```python
y_pred = dt_reg.predict(6.5)
y_pred
```

```python
dt_interp = np.arange(dataset[features].min(), dataset[features].max(), 0.1)
dt_interp = dt_interp.reshape(len(dt_interp), 1)
plt.scatter(dataset[features], dataset[target], c='r')
plt.plot(dt_interp, dt_reg.predict(dt_interp), c='g')
```

### Now use graphviz to visualize the tree

```python
from sklearn import tree
import pydotplus
from IPython.display import Image
dot_data = tree.export_graphviz(dt_reg, out_file=None,
                               feature_names=features,
                               class_names=target,
                               filled=True, rounded=True,
                               special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
#This line will show the graphical tree inline (for notebooks)
Image(graph.create_png())
#If you want to save the graphic
#graph.write_pdf('dt_reg.pdf')
#graph.write_png('dt_reg.png')
```
